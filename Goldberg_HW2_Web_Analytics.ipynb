```python
import networkx as nx
import matplotlib.pyplot as plot
%matplotlib inline
import numpy as np

g = nx.read_gml('./power/power.gml')
```

```python
#diameter of entire graph
nx.diameter(g)
```

```python
#cut the graph into a smaller graph
nodelist = range(0, 500)
g = g.subgraph(nodelist)
```

```python
#choose a connected graph from within
g1 = list(nx.connected_component_subgraphs(g))[0]
```

```python
#automated diameter of subgraph
nx.diameter(g1)
```

```python
#density: as expected with a randomly cut graph, the density is quite low
nx.density(g1)
```

```python
#distribution: few with 4+ connections in this cut-off network
deg = nx.degree(g1)
h=plot.hist(deg.values(),100)
```

```python
#average node degree: 2.5 average connections
np.mean(deg.values())
```

```python
#betweeness centrality: few key connectors
b = nx.betweenness_centrality(g1)

def sorted_map(map):
    ms = sorted(map.iteritems(), key=lambda (k,v): (-v,k)) 
    return ms

bs = sorted_map(b)
bs[:10]
```
