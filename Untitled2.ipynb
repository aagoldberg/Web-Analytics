```python
import networkx as nx
```

```python
hw1 = nx.Graph()
```

```python
hw1.add_nodes_from(["Andre", "Carol", "Fernando", "Diane", "Beverly", "Ed", "Garth", "Heather", "Ike", "Jane"])
```

```python
hw1.add_edges_from([("Andre", "Carol"), ("Andre", "Fernando"), ("Andre", "Diane"), ("Andre", "Beverly"), 
                   ("Beverly", "Diane"), ("Beverly", "Ed"), ("Beverly", "Garth"), 
                   ("Ed", "Diane"), ("Ed", "Garth"), 
                   ("Garth", "Fernando"), ("Garth", "Heather"),
                   ("Diane", "Carol"), ("Diane", "Fernando"),
                   ("Carol", "Fernando"), 
                   ("Fernando", "Heather"), 
                   ("Heather", "Ike"), 
                   ("Ike", "Jane")])
```

```python
hw1.graph()
```

```python
import matplotlib.pyplot as plt
```

```python
nx.draw(hw1)
```

```python
plt.show()
```
